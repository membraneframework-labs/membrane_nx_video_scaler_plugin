```python
import numpy as np
import matplotlib.pyplot as plt
```

### Data

```python
scaling_ratio = ["1/1.5", "1/2", "1/3", "1/4", "1/6"]

nx_avg_time = np.array([376.90, 289.97, 217.69, 139.05, 130.77])
ffmpeg_avg_time = np.array([255.43, 163.98, 139.26, 123.47, 99.80])
nx_to_ffmpeg_avg_time = nx_avg_time / ffmpeg_avg_time

nx_median_time = np.array([383.08, 288.74, 207.37, 136.04, 127.72])
ffmpeg_median_time = np.array([252.96, 255.70, 134.92, 119.23, 100.93])
nx_to_ffmpeg_median_time = nx_median_time / ffmpeg_median_time
```

### Charts

```python
fig, ax = plt.subplots(1, 2)
fig.set_size_inches(14, 6)
plt.prism()

ax[0].plot(np.array(scaling_ratio), np.array(nx_to_ffmpeg_avg_time))
ax[1].plot(np.array(scaling_ratio), np.array(nx_to_ffmpeg_median_time))

ax[0].set_title("Average time for Nx / average time for FFmpeg (ratio) for different scaling", size=12)
ax[1].set_title("Median time for Nx / median time for FFmpeg (ratio) for different scaling", size=12)

ax[0].set_xlabel("Scaling ratio - output size / input size", size=12)
ax[0].set_ylabel("Average time ratio", size=12)

ax[1].set_xlabel("Scaling ratio - output size / input size", size=12)
ax[1].set_ylabel("Median time ratio", size=12)

plt.tight_layout()

plt.show()
```
